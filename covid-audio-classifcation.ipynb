# Covid-19 audio classification

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Importing the dataset
df=pd.read_csv("./archive/metadata_compiled.csv")
df.shape
```

```python
# Deleting unlabelled rows for column status
df.dropna(subset=['status'],inplace=True)
df.shape
# Replacing NaN with 0
df.fillna(0,inplace=True)
df.shape
```

```python
# Adding path for the audio data in column called relative_path
df['relative_path']='/archive'+'/'+df['uuid'].astype(str)+'.wav'
df.shape
```

```python
for col in df.columns:
    print(col)
```

```python
age_fig=px.histogram(df, x='age', nbins=50, color_discrete_sequence=['purple'], opacity=0.5)
age_fig.show()
```

```python
gender_fig=px.histogram(df, x='gender', color='gender', pattern_shape='status')
gender_fig.show()
```
